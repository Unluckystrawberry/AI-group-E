```python
#preliminaries
import pandas as pd
import numpy as np

%matplotlib inline
import random
import matplotlib.pyplot as plt
```

```python
import feather
```

```python
df_1 = feather.read_dataframe("./covid_date.feather")
df_1
```

```python
df_2 = feather.read_dataframe("./fitbit_activities_fairly.feather")
df_2
```

# fitbit_activities_fairly

```python
date_merged_1and1 = pd.merge(df_1,df_2,on= 'user_id', how = 'left')
date_merged_1and1 = date_merged_1and1.iloc[:,:4]
```

```python
date_merged_1and1
```

```python
date_merged_1and1["date_illness"] = pd.to_datetime(date_merged_1and1["date_illness"]).dt.date
date_merged_1and1['datetime'] = date_merged_1and1['datetime'].dt.date
date_merged_1and1['difference'] = (date_merged_1and1['datetime'] - date_merged_1and1['date_illness']).dt.days
```

```python
date_merged_1and1
```

```python
date_merged_1and1_fairly = date_merged_1and1[(date_merged_1and1['difference'] > -8) & (date_merged_1and1['difference'] < 8)]
df_grouped_fairly = date_merged_1and1_fairly.groupby('user_id')
df_mean_fairly = df_grouped_fairly['activities_minutesFairlyActive'].mean()
df_result_fairly = df_mean_fairly.reset_index()
```

```python
df_result_fairly #The mean activities_minutes Fairly Active of the covod_19 people
```

```python
random_days = np.random.randint(60, 70, size=len(date_merged_1and1))
random_timedelta1 = pd.TimedeltaIndex(random_days, unit='d')
date_merged_1and1['non-Covid'] = date_merged_1and1['date_illness'] - random_timedelta1
date_merged_1and1['difference-non'] = (date_merged_1and1['datetime'] - date_merged_1and1['non-Covid']).dt.days
date_merged_1and1_fairly_non = date_merged_1and1[(date_merged_1and1['difference-non'] > -8) & (date_merged_1and1['difference-non'] < 8)]
df_grouped_fairly_non = date_merged_1and1_fairly_non.groupby('user_id')
df_mean_fairly_non = df_grouped_fairly_non['activities_minutesFairlyActive'].mean()
df_result_fairly_non = df_mean_fairly_non.reset_index()
```

```python
df_result_fairly_non #The mean activities_minutes Fairly Active of the non-covod_19 people
```
