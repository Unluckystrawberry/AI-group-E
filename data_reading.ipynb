```python
import pandas as pd
import datetime as datetime
import numpy as np
```

```python
date_merged_1and9 = pd.merge(df_1,df_9,on= 'user_id', how = 'left')
date_merged_1and9 = date_merged_1and9.iloc[:,:4]
date_merged_1and9["date_illness"] = pd.to_datetime(date_merged_1and9["date_illness"]).dt.date
date_merged_1and9['dateTime'] = date_merged_1and9['dateTime'].dt.date
date_merged_1and9['difference'] = (date_merged_1and9['dateTime'] - date_merged_1and9['date_illness']).dt.days
date_merged_1and9_rhr = date_merged_1and9[(date_merged_1and9['difference'] > -8) & (date_merged_1and9['difference'] < 8)]
df_grouped_rhr = date_merged_1and9_rhr.groupby('user_id')
df_mean_rhr = df_grouped_rhr['rhr'].mean()
df_result_rhr = df_mean_rhr.reset_index()
```

```python
df_result_rhr
```

```python
random_days = np.random.randint(60, 70, size=len(date_merged_1and9))
random_timedelta = pd.TimedeltaIndex(random_days, unit='d')
date_merged_1and9['non-Covid'] = date_merged_1and9['date_illness'] - random_timedelta
date_merged_1and9['difference-non'] = (date_merged_1and9['dateTime'] - date_merged_1and9['non-Covid']).dt.days
date_merged_1and9_rhr_non = date_merged_1and9[(date_merged_1and9['difference-non'] > -8) & (date_merged_1and9['difference-non'] < 8)]
df_grouped_rhr_non = date_merged_1and9_rhr_non.groupby('user_id')
df_mean_rhr_non = df_grouped_rhr_non['rhr'].mean()
df_result_rhr_non = df_mean_rhr_non.reset_index()
```

```python
gad7 = pd.merge(df_1,df_15,on= 'user_id', how = 'left')
gad7 = gad7.iloc[:,[0, 1, 6, 10]]
gad7["date_illness"] = pd.to_datetime(gad7["date_illness"]).dt.date
gad7["time_received"] = pd.to_datetime(gad7["time_received"]).dt.date
gad7['difference'] = (gad7['time_received'] - gad7['date_illness']).dt.days
Non_Covid_gad7 = gad7[gad7['difference'] > 0]
Covid_gad7 = gad7[gad7['difference'] <= 0]
Gad7_Covid_grouped = Covid_gad7.groupby('user_id')
Gad7_Covid_grouped_mean = Gad7_Covid_grouped['score'].mean()
Gad7_Covid_grouped_result = Gad7_Covid_grouped_mean.reset_index()
Gad7_Covid_grouped_result.rename(columns={'score': 'Gad7_Covid_score'}, inplace=True)
Gad7_non_Covid_grouped = Non_Covid_gad7.groupby('user_id')
Gad7_non_Covid_grouped_mean = Gad7_non_Covid_grouped['score'].mean()
Gad7_non_Covid_grouped_result = Gad7_non_Covid_grouped_mean.reset_index()
Gad7_non_Covid_grouped_result.rename(columns={'score': 'Gad7_non_Covid_score'}, inplace=True)
```
