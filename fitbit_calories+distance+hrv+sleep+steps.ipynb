```python
import feather
import pandas as pd
import numpy as np
import random
```

**covid_data**

```python
df = pd.read_feather('./dataset/covid_date.feather')
df
```

## calories ##

```python
df_cal= feather.read_dataframe("./dataset/fitbit_calories.feather")
df_cal
```

```python
df_cal_m = pd.merge(df,df_cal,on=['user_id'],how='left')
```

```python
df_cal_m
```

```python
df_cal_m["date_illness"] = pd.to_datetime(df_cal_m["date_illness"]).dt.date
df_cal_m['datetime'] = df_cal_m['datetime'].dt.date
df_cal_m['difference'] = (df_cal_m['datetime'] - df_cal_m['date_illness']).dt.days
```

```python
df_cal_m
```

```python
date_cal = df_cal_m[(df_cal_m['difference'] > -8) & (df_cal_m['difference'] < 8)]
df_grouped_cal = date_cal.groupby('user_id')
df_mean_cal = df_grouped_cal['activities_calories'].mean()
df_result_cal = df_mean_cal.reset_index()
```

```python
df_result_cal
#The mean activities_calories of the covod_19 people
```

```python
random_days = np.random.randint(60, 70, size=len(df_cal_m))
random_timedelta1 = pd.TimedeltaIndex(random_days, unit='d')
df_cal_m['non-Covid'] = df_cal_m['date_illness'] - random_timedelta1
df_cal_m['difference-non'] = (df_cal_m['datetime'] - df_cal_m['non-Covid']).dt.days
df_cal_non = df_cal_m[(df_cal_m['difference-non'] > -8) & (df_cal_m['difference-non'] < 8)]
df_grouped_cal_non = df_cal_non.groupby('user_id')
df_mean_cal_non = df_grouped_cal_non['activities_calories'].mean()
df_result_cal_non = df_mean_cal_non.reset_index()
```

```python
df_result_cal_non
#The mean activities_calories of the non-covod_19 people
```
